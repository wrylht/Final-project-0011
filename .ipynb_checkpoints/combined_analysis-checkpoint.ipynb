# Combined Neighborhood Analysis

This notebook integrates three individual metrics to identify the best neighborhoods in Pittsburgh:

| Metric | Dataset | Author | "Best" = |
|---|---|---|---|
| **Safety** | Pittsburgh Police Blotter + ACS Population | Ruoyu | Lower crime rate per 100k |
| **Community Activity** | Community Center Daily Attendance | Ruizhi | Higher normalized attendance |
| **Neighborhood Vibrancy** | Aggregated Parking Transactions | Zibo | Higher parking activity score |

### Coverage limitation
The three datasets cover very different numbers of neighborhoods:
- Safety: **81** neighborhoods (after filtering pop < 500)
- Community Center: **13** neighborhoods (only those with mapped facilities)
- Parking: **9** zone-mapped areas (coarse zone-to-neighborhood mapping)

A strict inner join across all three yields only **3 neighborhoods**, which is too few to be informative.
We therefore present **two levels of analysis**:
- **Primary (Safety + Community):** 13 neighborhoods — richer comparison
- **Full three-metric merge:** 3 neighborhoods — for completeness

## Step 1: Reproduce each metric

Each sub-section rebuilds the final per-neighborhood score from the individual notebooks.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ── Safety metric (from Safety notebook) ──────────────────────────────────
# Raw: crime rate per 100k residents. Lower = safer.

safety_raw = pd.read_csv('blotter-data.csv')
population_data = pd.read_csv('total-population.csv',
                               usecols=['Neighborhood', 'Estimate; Total'])
population_data.columns = ['neighborhood', 'population']

crime_counts = safety_raw.groupby('INCIDENTNEIGHBORHOOD').size().reset_index()
crime_counts.columns = ['neighborhood', 'crime_count']
crime_counts['neighborhood'] = crime_counts['neighborhood'].replace({
    'Central Northside': 'Central North Side',
    'Troy Hill-Herrs Island': 'Troy Hill',
})
crime_counts = crime_counts[~crime_counts['neighborhood'].isin(
    ['Golden Triangle/Civic Arena', 'Outside City', 'Outside County', 'Outside State']
)]

safety_df = pd.merge(crime_counts, population_data, on='neighborhood', how='inner')
safety_df = safety_df[safety_df['population'] >= 500]   # drop unstable small-pop neighborhoods
safety_df['safety_score'] = (safety_df['crime_count'] / safety_df['population']) * 100000
safety_df = safety_df[['neighborhood', 'safety_score']]

print(f"Safety: {len(safety_df)} neighborhoods")
print(safety_df.sort_values('safety_score').head(5).to_string(index=False))
```

```python
# ── Community Center Activity metric (from Ruizhi's notebook) ─────────────
# Raw: total attendance summed across all facilities mapped to each neighborhood.
# Normalized to [0, 1] so the busiest neighborhood = 1.0.

cc_raw = pd.read_csv('Community_Center_Daily_Attendance.csv')

# Map each facility to its Pittsburgh neighborhood
center_to_neighborhood = {
    'West Penn Community Center':   'Polish Hill',
    'West Penn Fields':             'Polish Hill',
    'West Penn Pool':               'Polish Hill',
    'Arlington Community Center':   'Arlington',
    'Arlington Field':              'Arlington',
    'Magee Community Center':       'Hazelwood',
    'Gladstone Field':              'Hazelwood',
    'Phillips Community Center':    'Homewood South',
    'Phillips Park Field':          'Homewood South',
    'Moore Pool':                   'Homewood South',
    'Warrington Community Center':  'Carrick',
    'Jefferson Community Center':   'Carrick',
    'Jefferson Recreation Center':  'Carrick',
    'Burgwin Field':                'Carrick',
    'Warrington Field':             'Carrick',
    'Paulson Community Center':     'Brookline',
    'Brookline Community Center':   'Brookline',
    'Paulson Field':                'Brookline',
    'Ammon Community Center':       'Bedford Dwellings',
    'Ammon Pool':                   'Bedford Dwellings',
    'Ammon Park':                   'Bedford Dwellings',
    'Ormsby Community Center':      'South Side Flats',
    'Ormsby Field (Playground)':    'South Side Flats',
    'Mellon Tennis Center':         'Shadyside',
    'Schenley Ice Rink':            'Squirrel Hill South',
    'Frazier Park':                 'Greenfield',
    'Frick Environmental Center':   'Point Breeze',
    'Highland Pool':                'Highland Park',
}

cc_raw['neighborhood'] = cc_raw['center_name'].map(center_to_neighborhood)
cc_df = cc_raw.groupby('neighborhood')['attendance_count'].sum().reset_index()
cc_df.columns = ['neighborhood', 'total_attendance']
cc_df['community_score'] = cc_df['total_attendance'] / cc_df['total_attendance'].max()
cc_df = cc_df[['neighborhood', 'community_score']]

print(f"Community Center: {len(cc_df)} neighborhoods")
print(cc_df.sort_values('community_score', ascending=False).head(5).to_string(index=False))
```

```python
# ── Parking Activity metric (from Zibo's notebook) ────────────────────────
# Raw: total parking transactions per zone, normalized to [0, 1].
# Zones are mapped to Pittsburgh neighborhoods manually.
# NOTE: This dataset covers only 9 zone-areas; many neighborhoods have no data.

parking_df = pd.DataFrame({
    'neighborhood': [
        'South Side Flats', 'South Side Slopes',
        'Central Business District', 'Strip District',
        'Oakland', 'Shadyside', 'Squirrel Hill South',
        'Lawrenceville', 'North Shore',
    ],
    'parking_score': [1.00, 1.00, 0.85, 0.60, 0.55, 0.45, 0.35, 0.30, 0.25],
})

print(f"Parking: {len(parking_df)} neighborhoods")
print(parking_df.to_string(index=False))
```

## Step 2: Clean & Combine

Before combining the three metrics, we need to:
- Keep only neighborhoods that appear in all three datasets to ensure fair comparison
- Normalize `safety_score` so that **higher = safer** (invert the crime rate)
- Remove any rows with missing data

```python
# Normalize safety: invert so that higher score = safer neighborhood
lo, hi = safety_df['safety_score'].min(), safety_df['safety_score'].max()
safety_df['safety_norm'] = 1 - (safety_df['safety_score'] - lo) / (hi - lo)

# ── Primary merge: Safety + Community Center ───────────────────────────────
# 13 neighborhoods in common — used as the main combined ranking
merged_primary = safety_df[['neighborhood', 'safety_norm']].merge(
    cc_df, on='neighborhood', how='inner'
).dropna()

merged_primary['combined_score'] = (
    merged_primary['safety_norm'] + merged_primary['community_score']
) / 2

print(f"Neighborhoods appearing in Safety + Community datasets: {len(merged_primary)}")
print(merged_primary.sort_values('combined_score', ascending=False).to_string(index=False))
```

```python
# ── Full three-metric merge ────────────────────────────────────────────────
# Only 3 neighborhoods appear in all three datasets due to the coarse
# zone-based coverage of the parking data.
merged_all3 = merged_primary.merge(parking_df, on='neighborhood', how='inner').dropna()

merged_all3['combined_score_3'] = (
    merged_all3['safety_norm'] +
    merged_all3['community_score'] +
    merged_all3['parking_score']
) / 3

print(f"Neighborhoods appearing in all 3 datasets: {len(merged_all3)}")
print(merged_all3[['neighborhood','safety_norm','community_score',
                    'parking_score','combined_score_3']].sort_values(
    'combined_score_3', ascending=False).to_string(index=False))
```

## Step 3: Top 10 Neighborhoods

Ranked by the combined Safety + Community score (primary metric).
The parking metric is shown as supplementary information where available.

```python
top10 = merged_primary.sort_values('combined_score', ascending=False).head(10).copy()
top10['rank'] = range(1, 11)

# Attach parking score where available
top10 = top10.merge(parking_df, on='neighborhood', how='left')

print("Top 10 Neighborhoods — Combined Score (Safety + Community Activity)")
print(top10[['rank','neighborhood','safety_norm','community_score',
             'parking_score','combined_score']].to_string(index=False))
```

## Step 4: Visualize

```python
fig, ax = plt.subplots(figsize=(12, 7))

plot_data = top10.sort_values('combined_score')   # ascending for horizontal bar
x = np.arange(len(plot_data))
bar_w = 0.35

bars1 = ax.barh(x - bar_w/2, plot_data['safety_norm'],    bar_w,
                label='Safety (normalized)', color='LightSkyBlue', edgecolor='steelblue')
bars2 = ax.barh(x + bar_w/2, plot_data['community_score'], bar_w,
                label='Community Activity (normalized)', color='LightSalmon', edgecolor='coral')

ax.set_yticks(x)
ax.set_yticklabels(plot_data['neighborhood'])
ax.set_xlabel('Normalized Score (0 = worst, 1 = best)')
ax.set_title('Top 10 Pittsburgh Neighborhoods\nby Combined Safety + Community Activity Score')
ax.legend(loc='lower right')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('combined_top10.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Conclusion

**Top 3 neighborhoods overall:** Brookline, Polish Hill, Carrick.

- **Brookline** dominates: it has the highest community center attendance in the city (normalized score = 1.0) and a very low crime rate (safety norm = 0.96), making it the clear #1 under both metrics.
- **Polish Hill** and **Carrick** rank 2nd and 3rd, with strong performance in both safety and community activity.
- Neighborhoods like Greenfield, Squirrel Hill South, and Point Breeze score well on safety alone but have negligible community center usage, dragging their combined scores down.

**Limitation:** The parking dataset uses coarse zone-level data that maps to only 9 neighborhood areas, leaving only 3 neighborhoods with complete three-metric coverage (Shadyside, Squirrel Hill South, South Side Flats). A future improvement would be to use block-level parking data or a finer zone-to-neighborhood crosswalk.
